```python
#IMPORTING IMPORTANT LIBRARIES
import numpy as np
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
```

```python
#READING THE DATA
data = pd.read_csv("dataset.csv")
```

```python
data
```

```python
#STEP-1
#CREATING A CONNECTOR TO THE DATABASE
csv_database = create_engine('sqlite:///csv_database.db')
```

```python
#STEP-2
#BUILDING THE DATABASE BY CHUNKSIZE
chunksize = 10000
i = 0
j = 0

for df in pd.read_csv("dataset.csv", chunksize=chunksize, iterator = True):#Coding the procedure for loading the data-set file iteration-after-iteration by chunksize
    df = df.rename(columns = {c: c.replace(' ','') for c in df.columns})#Removing spaces in column spaces
    df.index += j
    
    df.to_sql('data_use', csv_database, if_exists='append')#Appending the existing dataset by current(i) bunch of data
    j=df.index[-1]+1#Indexing the position for the next bunch of data(for next iteration)
    
    print('| index: {}'.format(j))
```

```python
#CONSTRUCTING THE DATAFRAME FROM THE DATABASE
#NOW WRITING THE REQUIRED QUERY
df = pd.read_sql_query('SELECT * FROM data_use WHERE TotalClaimAmount ', csv_database)
```
